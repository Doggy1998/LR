```python
import sympy as sp
```

```python
n = input('請問是幾維向量? ')

amount = 0
v_list = []

while True:
    keeping = input(f'按 Enter 開始輸入第{amount+1}個向量；離開請輸入 q\n')
    if keeping == '':
        v = []
        for i in range(int(n)):
            e = input(f'第{amount+1}個向量裡第{i+1}個元素')
            v.append(float(e))
        v_list.append(v)
        amount += 1
            
            
    elif (keeping == 'q') or (keeping=='Q'):
        break
    else:
        print('invalid input')
        continue
        

        
s = sp.Matrix(v_list).T

rref_m, pivot_c = s.rref()

from sympy import pprint
print(f'原矩陣:\n')
pprint(s)
print('\n\n')
print(f'pivot column(s):\n')
for c in pivot_c:
    pprint(s.col(c))
    print()
print('\n\n')
print('線性獨立向量: ')
for c in pivot_c:
    print(v_list[c])
print(f'最大線性獨立的數量: {len(pivot_c)}')
```
