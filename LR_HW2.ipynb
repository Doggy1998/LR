```python
## 詢問 equation

import numpy as np 

print('')
resourceDict = {}
# 限二維
# while True:

def getLimit(resourceDict):
    resource = input('請輸入原料')
    resourceAmount = input(f'請輸入{resource}的限有數量')
    resourceDict[resource] = float(resourceAmount)
    return resourceDict
    
    
productList = []
    
def getEquation(resourceDict, productList):
    
    productName = input('請輸入商品名稱: ')
    productList.append(productName)
    for resource in resourceDict:
        require = input(f'生產一單位{productName}需要幾單位的{resource}? ')
        try:
            requireArray = requireArray
            requireArray = np.append(requireArray, [[float(require)]], axis=0)
        except:
            requireArray = np.array([[float(require)]])
    
    return requireArray
    


    
inputResource = True
while inputResource:
    inputResource = input('是否輸入原料資訊? (y/n) ')
    if inputResource == 'y':
        resourceDict = getLimit(resourceDict)
    else:
        inputResource = False

inputProduct = True  

while inputProduct:
    inputProduct = input('是否輸入商品資訊? (y/n)')
    if inputProduct == 'y':
        require = getEquation(resourceDict, productList)
        try:
            # allRequireArray
            allRequireArray = np.append(allRequireArray, require, axis=1)
        except:
            allRequireArray = np.array(require)
    else:
        inputProduct = False
        
print('\n')

print(allRequireArray)

print('\n dot \n')

for p in productList:
    try:
        pArray = np.append(pArray, [[p]], axis=0)
    except:
        pArray = np.array([[p]])
    
print(pArray)

print('\n = \n')
for resource in resourceDict:
    try:
        rAmount = np.append(rAmount, [[resourceDict[resource]]], axis=0)
    except:
        rAmount = np.array([[resourceDict[resource]]]) 
print(rAmount)
    
    
del pArray
del rAmount
# del requireArray
del allRequireArray
```
